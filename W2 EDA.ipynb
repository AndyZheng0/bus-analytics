```python
#transform data. there is repeating objects. this will count each repeating data. 
df['region'] = df['region'].astype('category')
print(df['region'].value_counts())
df['control'] = df['control'].astype('category')
print(df['control'].value_counts())

```

```python
#this checks to see what is lined up,
#this is not complete, ignore  
col = ['control', 'gender', 'highest_degree', 'state']
df[col]= df[col].astype('category')
print(df[col].value_counts())
```

```python
#check missing values
df. isnull ().sum()
```

```python
#generate a bar chat
import seaborn as sns
ax = sns.countplot(data = df, x='region', palette= 'pastel', edgecolor='black', hue='control', legend=True)
#set the title and label
ax.set_title('distro of regions for student admin')
ax.set_xlabel('region')
ax.set_ylabel('freq')
ax.tick_params(axis='x', rotation = 45)
```

```python
from scipy.stats import pearsonr

corr, p_value = pearsonr(df['tuition'], df['median_debt'])


print ('Pearson Correlation: ' + str (corr))
print ('p-value: ' + str(p_value))
```
