## Cluster analysis




```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# for analysis
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering, KMeans
```

```python
#lead data
from sklearn import datasets
iris = datasets.load_iris()

```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.info()
```

```python
df. head()
```

```python
#check for missing data
df.isna().sum()
```

## visualizations


```python
corr_matrix = df [['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']].corr()
```

```python
sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm', vmin=-1, vmax= 1)
# closer to 1 or -1 means it is correlated, 1/-1 means its is literally the same value. CLoser to 0 means not correlated
# here, sepal length, petal length, and petal width are highly correlated
#--> only use sepal width and petal width in the analysis. 
```

```python
# Data prep
# scale the num var before cluster
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df.loc[:, ['sepal width (cm)', 'petal width (cm)']])
```

## Hierarchical Clustering

```python
# perform hierarchical Clustering using linkage
```

```python
linked = linkage(scaled_features, method = 'ward')
#dendrogram
plt.figure(figsize= (6, 4))
dend = dendrogram(linked, orientation ='top', distance_sort = 'descending', show_leaf_counts = True)
```

## hierarchical clustering

```python
# get the  cluster membership
hierarchical = AgglomerativeClustering (n_clusters = 3, linkage ='ward')
df['h_cluster'] = hierarchical.fit_predict(scaled_features)
df.info()
```

## k mean clustering

```python
kmeans = KMeans(n_clusters=3,random_state=11)
df['k_cluster'] = kmeans.fit_predict(scaled_features)
df.info()
```

```python
#export result
df.to_csv('cluster_result.csv')
```

## visualizations for optimal number of clusters


```python
# elbow method
wcss = []
cluster_range = range (1, 11) #test 1 - 10 cluster

for k in cluster_range:
    temp = KMeans(n_clusters = k, random_state = 11).fit(scaled_features)
    wcss.append(temp.inertia_)

# plot the graph
plt.plot(cluster_range, wcss, marker = 'o', linestyle = '-')
plt.grid(True)
```

```python
#silhouette method, higher is better, but not too low a number of groups
from sklearn.metrics import silhouette_score
silhouette_scores = []
cluster_range = range (2,11)

for k in cluster_range:
    temp= KMeans(n_clusters=k, random_state = 11).fit_predict(scaled_features)
    score= silhouette_score(scaled_features, temp)
    silhouette_scores.append(score)

plt.plot(cluster_range, silhouette_scores, marker='o', linestyle = '-')
plt.grid(True)
```

## profile the clusters 


```python
#profiling for hierarchical clustering
h_profiles = df.groupby('h_cluster').mean()
h_profiles
```

```python
#profiling for k-means clustering
k_profiles = df.groupby('k_cluster').mean()
k_profiles
```
