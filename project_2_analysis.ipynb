## EDA

```python
df = pd.read_csv('data/IL_business.csv')
df2 = pd.read_csv('data/IL_review.csv')

merged=pd.merge(df, df2, on='business_id', how='inner')
merged.head()

```

```python
corr_matrix = merged[['stars_x', 'stars_y','review_count','useful','funny','cool']].corr()
```

```python
sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm', vmin=-1, vmax= 1)
```
